```python
from math import gcd

# https://www.youtube.com/@EricRowland

def R_generator(start):
    this = ( 0, 0, 0, False, False )
    
    R = start
    n = 1
    g = 0

    cluster_cnt = 0
    in_cluster = False
    
    while True:
        prev = this
        is_prime = n*3 == R

        if not in_cluster and is_prime:
            in_cluster = True
            cluster_cnt = cluster_cnt + 1

        if in_cluster:
            if g == 1:
                in_cluster = False

        this = ( n, R, g, is_prime, cluster_cnt if in_cluster else False )
        yield this
        
        n += 1
        g = gcd(n, R)
        R = R + g
```
